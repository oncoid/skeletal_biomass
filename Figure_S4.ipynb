```python
import pandas as pd 
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import utils
```

```python
data_file = './data/paleo_meso_ceno_data_paleogeography_epoch.csv'

data = pd.read_csv(data_file)
data = data.sort_values(by=['mean_ma'], ascending=False)

time_file = './data/timeScale.xlsx'
time = pd.read_excel(time_file,sheet_name='Sheet1')
time_period = utils.get_timesubset(time,3,541)
width_period = [time_period['max_ma'][i] - time_period['min_ma'][i] for i in list(time_period.index)]

time_epoch = utils.get_timesubset(time,4,541)
width_epoch = [time_epoch['max_ma'][i] - time_epoch['min_ma'][i] for i in list(time_epoch.index)]

time_age = utils.get_timesubset(time,5,541)
width_age = [time_age['max_ma'][i] - time_age['min_ma'][i] for i in list(time_age.index)]
```

```python
data_skeletal = utils.get_stats(data,'skeletal_total','epoch')
data_skeletal = data_skeletal.sort_values(by=['time'], ascending=True)
```

```python
fig = make_subplots(rows = 1,cols = 1)

fig.add_trace(go.Histogram(x=data.epoch,y=data.numberOfSamples,
                          showlegend=False,marker_color='grey'),row=1,col=1)

#Llandovery
fig.add_annotation(x=7, y=12,
            text="n=1",
            showarrow=False,yshift=10)


fig.update_xaxes(ticks = 'outside',showticklabels = True,showline = True, tickangle=-60,
                     mirror = True,linecolor = 'black',row=1,col=1)
fig.update_yaxes(title = 'Number of samples',ticks = 'outside',showline = True, range=(0,755),
                         mirror = True,linecolor = 'black',row=1,col=1)

fig.update_layout(height = 600, width = 1000,font_color = 'black',
                  font_family='Helvetica', font_size=14,
                 paper_bgcolor = 'white',
                 plot_bgcolor = 'white')

fig.show()
fig.write_image('./figures/Figure_S4.png',scale=6)
```
